```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.datasets import load_iris

# Load the Iris dataset (a classic example for data visualization)
iris = load_iris()
X = iris.data
y = iris.target

# Apply t-SNE to reduce the data to 2D for visualization
tsne = TSNE(n_components=2, random_state=0)
X_2D = tsne.fit_transform(X)

# Plot the 2D t-SNE representation
plt.figure(figsize=(8, 6))
plt.scatter(X_2D[:, 0], X_2D[:, 1], c=y, cmap=plt.cm.get_cmap("Set1", 3))
plt.title("t-SNE Visualization of Iris Dataset")
plt.colorbar(ticks=range(3), label="Iris Classes")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

# Create a toy dataset in a high-dimensional space (6D for simplicity)
np.random.seed(0)
n_samples = 300
n_features = 6
X = np.random.randn(n_samples, n_features)

# Create three different shapes of data
circle = np.random.randn(n_samples // 3, n_features) + [3, 0, 0, 0, 0, 0]
triangle = np.random.randn(n_samples // 3, n_features) + [0, 3, 0, 0, 0, 0]
square = np.random.randn(n_samples // 3, n_features) + [0, 0, 3, 0, 0, 0]

X = np.vstack([circle, triangle, square])
y = np.array([0] * (n_samples // 3) + [1] * (n_samples // 3) + [2] * (n_samples // 3))

# Apply t-SNE to reduce the data to 2D for visualization
tsne = TSNE(n_components=2, random_state=0)
X_2D = tsne.fit_transform(X)

# Plot the 2D t-SNE representation
plt.figure(figsize=(8, 6))
plt.scatter(X_2D[y == 0, 0], X_2D[y == 0, 1], label="Circle")
plt.scatter(X_2D[y == 1, 0], X_2D[y == 1, 1], label="Triangle")
plt.scatter(X_2D[y == 2, 0], X_2D[y == 2, 1], label="Square")
plt.legend()
plt.title("t-SNE Visualization of Different Shapes")
plt.show()
```

DBSCAN

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs

# Generate synthetic data with three clusters
data, labels = make_blobs(n_samples=300, centers=3, cluster_std=0.6, random_state=0)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
clusters = dbscan.fit_predict(data)

# Visualize the clustering results
plt.figure(figsize=(8, 6))
plt.scatter(data[:, 0], data[:, 1], c=clusters, cmap='viridis')
plt.title("DBSCAN Clustering")
plt.show()
```
